Import the corresponding dependencies: NumPy, OpenCV, and Matplotlib:

```python
import numpy as np #Numpy
import cv2
#OpenCV
from matplotlib import pyplot as plt #Matplotlib
count = 0
```

we are going to process in grayscale using OpenCV and show using Matplotlib:

```python
img = cv2.imread('numbers.png',0)
plt.imshow(img,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.show()
```

If you want to recognize those digits using machine learning or any other
algorithm, you need to simplify the visualization of them. Using thresholding
seems to be the first logical step to proceed with this exercise. We have learned
some thresholding methods, but the most commonly used one is Otsu's
binarization, as it automatically calculates the threshold value without the user
providing the details manually.

```python
#Apply Otsu's binarization to the grayscale image and show it using Matplotlib:
_,th1=cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU )
th1 = (255-th1)
# This step changes the black with white and vice versa in order to have white figures
plt.imshow(th1,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.show()
```

In order to get rid of the lines in the background, we need to do some
morphological transformations. First, start by applying the closing method:

```python
open1 = cv2.morphologyEx(th1, cv2.MORPH_OPEN, np.ones((4, 4),np.uint8))
plt.imshow(open1,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.show()
```

The lines in the background have been removed completely. Now a number
prediction will be much easier.

In order to fill the holes that are visible in these digits, we need to apply the
opening method. Apply the opening method to the preceding image:

```python
close1 = cv2.morphologyEx(open1, cv2.MORPH_CLOSE, np.ones((8, 8),
np.uint8))
plt.imshow(close1,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.show()
```

There are still leftovers and imperfections around the digits. In order to remove
these, a closing method with a bigger kernel would be the best choice. Now apply
the corresponding method:

```python
open2 = cv2.morphologyEx(close1, cv2.MORPH_OPEN,np.ones((7,12),np.uint8))
plt.imshow(open2,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.show()
```
